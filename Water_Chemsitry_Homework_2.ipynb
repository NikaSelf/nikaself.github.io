```python
import matplotlib.pyplot as plt
import numpy as np

# pH range
pH = np.linspace(0, 14, 500)
H = 10**(-pH)

# pKa values
pKa_H3AsO4 = [2.20, 7.01, 11.80]
pKa_H3AsO3 = [9.17, 14.10]

ka1, ka2, ka3 = [10**(-p) for p in pKa_H3AsO4]
ka4, ka5 = [10**(-p) for p in pKa_H3AsO3]

# Triprotic arsenic acid H3AsO4
denom1 = H**3 + H**2*ka1 + H*ka1*ka2 + ka1*ka2*ka3
alpha_H3AsO4 = H**3 / denom1
alpha_H2AsO4 = H**2 * ka1 / denom1
alpha_HAsO4  = H * ka1 * ka2 / denom1
alpha_AsO4   = ka1 * ka2 * ka3 / denom1

# Diprotic arsenous acid H3AsO3
denom2 = H**2 + H*ka4 + ka4*ka5
alpha_H3AsO3 = H**2 / denom2
alpha_H2AsO3 = H * ka4 / denom2
alpha_HAsO3  = ka4 * ka5 / denom2

# -------------------------------
# Plot 1: Arsenic acid H3AsO4
# -------------------------------
plt.figure(figsize=(10,6))

plt.plot(pH, alpha_H3AsO4, label="H₃AsO₄")
plt.plot(pH, alpha_H2AsO4, label="H₂AsO₄⁻")
plt.plot(pH, alpha_HAsO4, label="HAsO₄²⁻")
plt.plot(pH, alpha_AsO4,  label="AsO₄³⁻")

# Add vertical dashed lines at relevant pKa values
for pk in pKa_H3AsO4:
    plt.axvline(pk, color="k", linestyle="--", alpha=0.6)
    plt.text(pk+0.1, 0.9, f"pKa={pk}", rotation=90, va="center", fontsize=9)

plt.xlabel("pH")
plt.ylabel("Fraction (α)")
plt.title("Speciation of Arsenic Acid (H₃AsO₄)")
plt.legend()
plt.grid(True, linestyle=":")
plt.ylim(0, 1.05)
plt.xlim(0, 14)

plt.show()

# -------------------------------
# Plot 2: Arsenous acid H3AsO3
# -------------------------------
plt.figure(figsize=(10,6))

plt.plot(pH, alpha_H3AsO3, label="H₃AsO₃")
plt.plot(pH, alpha_H2AsO3, label="H₂AsO₃⁻")
plt.plot(pH, alpha_HAsO3, label="HAsO₃²⁻")

# Add vertical dashed lines at relevant pKa values
for pk in pKa_H3AsO3:
    plt.axvline(pk, color="k", linestyle="--", alpha=0.6)
    plt.text(pk+0.1, 0.9, f"pKa={pk}", rotation=90, va="center", fontsize=9)

plt.xlabel("pH")
plt.ylabel("Fraction (α)")
plt.title("Speciation of Arsenous Acid (H₃AsO₃)")
plt.legend()
plt.grid(True, linestyle=":")
plt.ylim(0, 1.05)
plt.xlim(0, 14)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# pH range
pH = np.linspace(0, 14, 500)
H = 10**(-pH)

# pKa values
pKa_H3AsO4 = [2.20, 7.01, 11.80]
pKa_H3AsO3 = [9.17, 14.10]

ka1, ka2, ka3 = [10**(-p) for p in pKa_H3AsO4]
ka4, ka5 = [10**(-p) for p in pKa_H3AsO3]

# Triprotic arsenic acid H3AsO4
denom1 = H**3 + H**2*ka1 + H*ka1*ka2 + ka1*ka2*ka3
alpha_H3AsO4 = H**3 / denom1
alpha_H2AsO4 = H**2 * ka1 / denom1
alpha_HAsO4  = H * ka1 * ka2 / denom1
alpha_AsO4   = ka1 * ka2 * ka3 / denom1

# Diprotic arsenous acid H3AsO3
denom2 = H**2 + H*ka4 + ka4*ka5
alpha_H3AsO3 = H**2 / denom2
alpha_H2AsO3 = H * ka4 / denom2
alpha_HAsO3  = ka4 * ka5 / denom2

# -------------------------------
# Plot 1: Arsenic acid H3AsO4
# -------------------------------
plt.figure(figsize=(10,6))

plt.plot(pH, alpha_H3AsO4, label="H₃AsO₄")
plt.plot(pH, alpha_H2AsO4, label="H₂AsO₄⁻")
plt.plot(pH, alpha_HAsO4, label="HAsO₄²⁻")
plt.plot(pH, alpha_AsO4,  label="AsO₄³⁻")

# Add vertical dashed lines at relevant pKa values
for pk in pKa_H3AsO4:
    plt.axvline(pk, color="k", linestyle="--", alpha=0.6)
    plt.text(pk+0.1, 0.9, f"pKa={pk}", rotation=90, va="center", fontsize=9)

# Shade pH 6.5–8.5 region
plt.axvspan(6.5, 8.5, color="gray", alpha=0.2)

plt.xlabel("pH")
plt.ylabel("Fraction (α)")
plt.title("Speciation of Arsenic Acid (H₃AsO₄)")
plt.legend()
plt.grid(True, linestyle=":")
plt.ylim(0, 1.05)
plt.xlim(0, 14)

plt.show()

# -------------------------------
# Plot 2: Arsenous acid H3AsO3
# -------------------------------
plt.figure(figsize=(10,6))

plt.plot(pH, alpha_H3AsO3, label="H₃AsO₃")
plt.plot(pH, alpha_H2AsO3, label="H₂AsO₃⁻")
plt.plot(pH, alpha_HAsO3, label="HAsO₃²⁻")

# Add vertical dashed lines at relevant pKa values
for pk in pKa_H3AsO3:
    plt.axvline(pk, color="k", linestyle="--", alpha=0.6)
    plt.text(pk+0.1, 0.9, f"pKa={pk}", rotation=90, va="center", fontsize=9)

# Shade pH 6.5–8.5 region
plt.axvspan(6.5, 8.5, color="gray", alpha=0.2)

plt.xlabel("pH")
plt.ylabel("Fraction (α)")
plt.title("Speciation of Arsenous Acid (H₃AsO₃)")
plt.legend()
plt.grid(True, linestyle=":")
plt.ylim(0, 1.05)
plt.xlim(0, 14)

plt.show()
```
